```python
#find pdfs have x pages
def find_pdfs_has_x_pages(tika_data, pagecount):
  #df.loc[df['id'] == 123, ['id','fname','lname']]
  tika_data_has_x_pages = tika_data.loc[tika_data['total_page_count'] == pagecount, ['file_path']]
  #tika_data_has_x_pages = tika_data.query("total_page_count == page_count")[['file_path']]
  return tika_data_has_x_pages
```

```python
#take file path list of pdf files has x page(s)
def take_file_paths_list(tika_data, pagecount):
  tika_data_has_x_pages = find_pdfs_has_x_pages(tika_data ,pagecount)
  x_page_pdfs_from_tika = []
  for tika_datum in tika_data_has_x_pages['file_path']:
    file_name = tika_datum.split('/')[-1].split('.')[0]
    print(file_name)
    x_page_pdfs_from_tika.append(file_name)
  return x_page_pdfs_from_tika
  #file_name = tika_data['file_path'].split('/')
```

```python
tika_file_paths_has_1_page = take_file_paths_list(tika_data = tika_data, pagecount = 1)
tika_file_paths_has_2_page = take_file_paths_list(tika_data = tika_data, pagecount = 2)
tika_file_paths_has_3_page = take_file_paths_list(tika_data = tika_data, pagecount = 3)
tika_file_paths_has_4_page = take_file_paths_list(tika_data = tika_data, pagecount = 4)
tika_file_paths_has_5_page = take_file_paths_list(tika_data = tika_data, pagecount = 5)
tika_file_paths_has_6_page = take_file_paths_list(tika_data = tika_data, pagecount = 6)
tika_file_paths_has_7_page = take_file_paths_list(tika_data = tika_data, pagecount = 7)
tika_file_paths_has_8_page = take_file_paths_list(tika_data = tika_data, pagecount = 8)
tika_file_paths_has_9_page = take_file_paths_list(tika_data = tika_data, pagecount = 9)
tika_file_paths_has_10_page = take_file_paths_list(tika_data = tika_data, pagecount = 10)
tika_file_paths_has_20_page = take_file_paths_list(tika_data = tika_data, pagecount = 20)
```
